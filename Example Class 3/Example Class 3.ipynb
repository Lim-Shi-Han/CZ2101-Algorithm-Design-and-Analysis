## Recursive definition: P(C) = max(P(C-iw$_j$, j-1) + ip$_j$, P(C, j-1)), where 0 <= i <= floor(C/w$_j$)

```python
def knapsack(w, p, C, n):
    profit = [[0 for c in range(n+1)] for r in range(C+1)] #set the entire 2d array to 0
    for r in range(1,C+1): #traverse every row and column
        for c in range(1,n+1): 
            profit[r][c] = profit[r][c-1] #assume the object in this column isn't added
            noOfObjects = 0
            while True:
                if(noOfObjects*w[c-1] <= r): #account for more than 1 instance of the object in the knapsack
                    if(profit[r][c] < profit[r-noOfObjects*w[c-1]][c-1] + noOfObjects*p[c-1]): #replace maximum profit
                        profit[r][c] = profit[r-noOfObjects*w[c-1]][c-1] + noOfObjects*p[c-1]
                    noOfObjects+=1
                else:
                    break
    return profit[C][n]
```

```python
w = [4, 6, 8]
p = [7, 6, 9]
print(knapsack(w,p,14,3))
```
